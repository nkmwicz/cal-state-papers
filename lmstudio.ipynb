```python
import lmstudio as lms

model = lms.llm("google/gemma-3-4b")
result = model.respond("What is the capital of France?")
print(result)
```

```python
from lmstudio import BaseModel


class ResSchema(BaseModel):
    capital: str
    latitude: float
    longitude: float


res = model.respond("Tell me about the capital of France.", response_format=ResSchema)
res = res.parsed
print(res)
```
